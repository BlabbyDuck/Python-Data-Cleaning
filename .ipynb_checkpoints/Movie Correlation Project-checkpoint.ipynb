```python
#Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import re as re

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# read data
df1 = pd.read_csv(r'/Users/blabbyduck/Desktop/OwnProj/Python data cleaning/movies.csv')
```

```python
#Look at data
df1.head()
```

```python
#checking for missing outputs
df1.isnull().sum().sum()
for col in df1.columns:
    percent_missing = np.mean(df1[col].isnull())
    print('{} - {}%'.format(col, percent_missing))
```

```python
#sum of null value
df1.isnull().sum().sum()
```

```python
df = df1.dropna()

for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, percent_missing))
```

```python
#Data type

df.dtypes
```

```python
#Change data type from float to int
df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
def find_year(text):
    result = re.findall(r"\d{4}", text)
    return result[0]
```

```python
#creating a new column named year
df["yearcorrected"] = df['released'].apply(lambda x: find_year(x))

df.head()
```

```python
#Sort the data by gross
df = df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
df.head()
```

```python
#scatter plot to compare budget and gross_revenue. Hypothesis: higher the budget, higher the gross_revenue
plt.scatter(x = df["budget"], y = df["gross"])
plt.title('Budget Vs Gross Earning')
plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')
plt.show
```

```python
df.head()
```

```python
#REGRESSION PLOT

sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws= {'color' : 'red'}, line_kws = {"color" : "blue"})
```

```python
#Correlation
df.corr(method = 'pearson') #Pearson(default), kendall, spearman
#df.corr(method = 'kendall')
#df.corr(method = 'spearman')
```

```python
#High correlation btwn Budget and Gross
```

```python
correlation_matrix = df.corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation Matrix for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()
```

```python
#Give each company a unique ID

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
#Correlation of every single field

correlation_matrix = df_numerized.corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation Matrix for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()
```

```python
df_numerized.corr()
```

```python
#to see correlation quickly
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
#even cleaner
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
